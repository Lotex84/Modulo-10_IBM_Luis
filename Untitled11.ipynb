```python
# Importar librerías necesarias
import pandas as pd
import numpy as np

# Paso 1: Cargar el dataset
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv"
df = pd.read_csv(url)

# Mostrar las primeras filas para inspección
print("Primeras filas del dataset:")
print(df.head(10))

# Paso 2: Revisar valores faltantes
missing_values = df.isnull().sum() / len(df) * 100
print("\nPorcentaje de valores faltantes por columna:")
print(missing_values)

# Paso 3: Revisar tipos de datos
print("\nTipos de datos en el dataset:")
print(df.dtypes)

# Paso 4: Análisis de la columna LaunchSite
launch_site_counts = df['LaunchSite'].value_counts()
print("\nNúmero de lanzamientos por sitio de lanzamiento (LaunchSite):")
print(launch_site_counts)

# Paso 5: Análisis de la columna Orbit
orbit_counts = df['Orbit'].value_counts()
print("\nNúmero de lanzamientos por tipo de órbita (Orbit):")
print(orbit_counts)

# Paso 6: Análisis de resultados de aterrizaje
landing_outcomes = df['Outcome'].value_counts()
print("\nResultados de aterrizaje:")
print(landing_outcomes)

# Paso 7: Filtrar resultados de aterrizaje no exitosos
# Identificar malos resultados
bad_outcomes = set(['Failure (drone ship)', 'Failure (ground pad)', 'No attempt', 'Uncontrolled', 'Precluded (drone ship)'])
failed_landings = df[df['Outcome'].isin(bad_outcomes)].shape[0]
print(f"\nNúmero de aterrizajes fallidos: {failed_landings}")
```

```python
import requests
import pandas as pd
import numpy as np
import datetime

# Configuración para mostrar todas las columnas y datos completos
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

# URL de la API de SpaceX
spacex_url = "https://api.spacexdata.com/v4/launches/past"

# Obtener datos desde la API de SpaceX
response = requests.get(spacex_url)
if response.status_code == 200:
    data = response.json()
else:
    print("Error al obtener los datos de la API")
    data = []

# Normalizar los datos
df_normalized = pd.json_normalize(data)

# Funciones para obtener datos de columnas específicas
def getBoosterVersion(data):
    """Obtiene las versiones del cohete."""
    booster_versions = []
    for x in data['rocket']:
        if x:
            response = requests.get(f"https://api.spacexdata.com/v4/rockets/{x}").json()
            booster_versions.append(response['name'])
    return booster_versions

def getLaunchSite(data):
    """Obtiene los sitios de lanzamiento."""
    longitude, latitude, launch_site = [], [], []
    for x in data['launchpad']:
        if x:
            response = requests.get(f"https://api.spacexdata.com/v4/launchpads/{x}").json()
            longitude.append(response['longitude'])
            latitude.append(response['latitude'])
            launch_site.append(response['name'])
    return longitude, latitude, launch_site

# Verificar el año en la primera fila de 'static_fire_date_utc'
if 'static_fire_date_utc' in df_normalized.columns:
    first_date = df_normalized['static_fire_date_utc'].iloc[0]
    year = pd.to_datetime(first_date).year
    print("El año en la primera fila es:", year)
else:
    print("Columna 'static_fire_date_utc' no encontrada.")

# Subset de datos
try:
    data_subset = df_normalized[['rocket', 'payloads', 'launchpad', 'cores', 'flight_number', 'date_utc']]
    # Procesamiento adicional
    data_subset['date'] = pd.to_datetime(data_subset['date_utc']).dt.date
    data_subset = data_subset[data_subset['date'] <= datetime.date(2020, 11, 13)]
    print("Datos procesados correctamente.")
except KeyError as e:
    print(f"Error procesando datos: {e}")
```

```python
import requests
import pandas as pd

# URL de la API de SpaceX
spacex_url = "https://api.spacexdata.com/v4/launches/past"

# Obtener datos desde la API de SpaceX
response = requests.get(spacex_url)
if response.status_code == 200:
    data = response.json()
else:
    print("Error al obtener los datos de la API")
    data = []

# Normalizar los datos
df_normalized = pd.json_normalize(data)

# Verificar si la columna 'cores' está presente
if 'cores' in df_normalized.columns:
    # Extraer la información del landingPad
    landing_pads = []
    for core in df_normalized['cores']:
        if core and isinstance(core, list) and len(core) > 0:
            landing_pads.append(core[0].get('landpad', None))
        else:
            landing_pads.append(None)
    
    # Agregar la columna 'landingPad' al DataFrame
    df_normalized['landingPad'] = landing_pads
    
    # Contar valores faltantes
    missing_values = df_normalized['landingPad'].isna().sum()
    print("Cantidad de valores que faltan en la columna 'landingPad':", missing_values)
else:
    print("Columna 'cores' no encontrada en los datos.")
```
